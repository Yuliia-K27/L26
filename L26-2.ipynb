```python
import matplotlib.pyplot as plt
import numpy as np

# Вихідні дані
x = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

# Побудова дискретного статистичного розподілу
unique, counts = np.unique(x, return_counts=True)
freq = counts / len(x)
distr = dict(zip(unique, freq))
print("Дискретний статистичний розподіл:")
print(distr)

# Побудова полігону частот
plt.figure(figsize=(8,6))
plt.hist(x, bins=np.arange(5.5, 15.5, 1), rwidth=0.8, alpha=0.7, density=True, edgecolor='black')
plt.xlabel('X')
plt.ylabel('Частота')
plt.title('Полігон частот')
plt.show()

# Побудова F*(x)
keys = list(distr.keys())
values = list(distr.values())
cumulative = np.cumsum(values)
plt.figure(figsize=(8,6))
plt.step(keys, cumulative, where='post', label='F*(x)')
plt.plot(np.concatenate(([keys[0]], keys)), [0] + cumulative.tolist(), 'ro')
plt.xlabel('X')
plt.ylabel('F*(x)')
plt.title('F*(x)')
plt.legend()
plt.show()

# Знаходження моди
mode = keys[np.argmax(values)]
print(f"Мода: {mode}")

# Знаходження медіани
x_sorted = sorted(x)
n = len(x_sorted)
if n % 2 == 0:
    median = (x_sorted[n//2-1] + x_sorted[n//2]) / 2
else:
    median = x_sorted[n//2]
print(f"Медіана: {median}")
```
